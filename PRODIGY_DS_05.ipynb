```python
#handling duplicates
df.duplicated().sum()
```

```python
#removing duplicates
df.drop_duplicates(inplace=True)
```

```python
# Convert 'Start_Time' to datetime format
df['Start_Time'] = pd.to_datetime(df['Time'])

# Extract hour and day of the week
df['Hour'] = df['Start_Time'].dt.hour
df['DayOfWeek'] = df['Start_Time'].dt.day_name()

# Plot accidents by hour
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Hour', palette="coolwarm")
plt.title("Accidents by Hour of the Day")
plt.xlabel("Hour")
plt.ylabel("Accident Count")
plt.show()
```

```python
# Plot accidents by weather condition
plt.figure(figsize=(12,5))
# Check if 'Weather_Condition' is in the DataFrame columns
if 'Weather_Condition' in df.columns:
    sns.countplot(y=df['Weather_Condition'], order=df['Weather_Condition'].value_counts().index[:10], palette="viridis")
    plt.title("Top 10 Weather Conditions During Accidents")
    plt.xlabel("Accident Count")
    plt.ylabel("Weather Condition")
    plt.show()
else:
    print("Column 'Weather_Condition' not found in DataFrame.")
    # You can print the available columns for inspection
    print("Available columns:", df.columns)
```

```python
# Sample 5000 rows for faster processing, or the maximum available if less than 5000
# Assuming 'Start_Lat' and 'Start_Lng' are incorrect and the actual columns are 'Latitude' and 'Longitude'
df_geo = df[['Latitude', 'Longitude']].dropna()
sample_size = min(5000, len(df_geo))  # Get the smaller value between 5000 and the number of rows in df_geo
df_sample = df_geo.sample(n=sample_size, random_state=42)

# Create a base map
import folium # Make sure to import folium
map_accidents = folium.Map(location=[df_sample["Latitude"].mean(), df_sample["Longitude"].mean()], zoom_start=5)

# Add points to the map
for lat, lng in zip(df_sample["Latitude"], df_sample["Longitude"]):
    folium.CircleMarker([lat, lng], radius=2, color="red", fill=True).add_to(map_accidents)

# Save and display the map
map_accidents.save("accident_hotspots.html")
map_accidents
```
